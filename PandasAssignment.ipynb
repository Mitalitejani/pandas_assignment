## Ex 1 Getting and Knowing your Data


```python
# Step 1. Import the necessary libraries

import pandas as pd
```

```python
# Step 2. Import the dataset from this address.

chipo = pd.read_csv('chipotle.csv')
```

```python
# Step 3. Assign it to a variable called chipo.

chipo
```

```python
# Step 4. See the first 10 entries

chipo.head(10)
```

```python
# Step 5. What is the number of observations in the dataset?

row_count = chipo.shape[0]
print('Number of observations in dataset :',row_count)
```

```python
# Step 6. What is the number of columns in the dataset?

col_count = len(chipo.columns)
print("Number of columns in dataset :",col_count)
```

```python
# Step 7. Print the name of all the columns.

chipo.columns
```

```python
# Step 8. How is the dataset indexed?

index_type = type(chipo.index)
print("Type of index:", index_type)
```

```python
# Step 9. Which was the most-ordered item?

item_orders = chipo.groupby('item_name')['quantity'].sum()

most_ordered_itemname = item_orders.idxmax()
print("Most-ordered item name:", most_ordered_itemname)
```

```python
# Step 10. For the most-ordered item, how many items were ordered?

item_orders = chipo.groupby('item_name')['quantity'].sum()

max_quantity_ordered = item_orders.max()
print("Quantity ordered for the most-ordered item:", max_quantity_ordered)
''
```

```python
# Step 11. What was the most ordered item in the choice_description column?

choice_orders = chipo.groupby('choice_description')['quantity'].sum()

most_ordered_choice = choice_orders.idxmax()
max_quantity_ordered_choice = choice_orders.max()

print("Most-ordered item in the choice_description column:", most_ordered_choice)
print("Total quantity ordered for the most-ordered choice:", max_quantity_ordered_choice)
```

```python
# Step 12. How many items were orderd in total?

total_ordered = chipo['quantity'].sum()

print("Total number of items ordered:", total_ordered)
```

```python
# Step 13. Turn the item price into a float

chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)
```

```python
# Step 13.a. Check the item price type

print(chipo['item_price'].dtype)
```

```python
# Step 13.b. Create a lambda function and change the type of item price

# convert_to_float = lambda x: float(x.strip('$'))
# chipo['item_price'] = chipo['item_price'].apply(convert_to_float)
```

```python
# Step 13.c. Check the item price type

# print(chipo['item_price'].dtype)
```

```python
# Step 14. How much was the revenue for the period in the dataset

chipo['Revenue'] = chipo['quantity'] * chipo['item_price']
total_revenue = chipo['Revenue'].sum()

print("Total revenue for the period:", total_revenue)
```

```python
# Step 15. How many orders were made in the period?

num_orders = chipo['order_id'].nunique()

print("Total number of orders made in the period:", num_orders)
```

```python
# Step 16. What is the average revenue amount per order?

average_revenue_per_order = total_revenue / num_orders

print("Average revenue amount per order:", average_revenue_per_order)
```

```python
# Step 17. How many different items are sold?

num_unique_items = chipo['item_name'].nunique()

print("Number of different items sold:", num_unique_items)
```

# Ex - GroupBy

```python
# Step 1. Import the necessary libraries

import pandas as pd
```

```python
#Step 2. Import the dataset from this address.

drinks = pd.read_csv('drinks.csv')
```

```python
# Step 3. Assign it to a variable called drinks.

drinks
```

```python
# Step 4. Which continent drinks more beer on average?

drinks.groupby('continent').beer_servings.mean()
```

```python
# Step 5. For each continent print the statistics for wine consumption.

drinks.groupby('continent').wine_servings.describe()
```

```python
# Step 6. Print the mean alcohol consumption per continent for every column

drinks.groupby('continent').mean()
```

```python
# Step 7. Print the median alcohol consumption per continent for every column

drinks.groupby('continent').median()
```

```python
# Step 8. Print the mean, min and max values for spirit consumption.

drinks.groupby('continent').spirit_servings.agg(['mean', 'min', 'max'])
```
